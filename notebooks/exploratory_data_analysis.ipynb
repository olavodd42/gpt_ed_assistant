```python
import os
import pandas as pd
import numpy as np

dataset_path = '../data/raw/processed/'
df_train = pd.read_csv(os.path.join(dataset_path, 'train.csv'))
df_val = pd.read_csv(os.path.join(dataset_path, 'test.csv'))
```

```python
from datetime import datetime

def calculate_stay_time(df, fmt):
    df['stay_time'] = df['outtime'].apply(lambda x: datetime.strptime(x, fmt)) - df['intime'].apply(lambda x: datetime.strptime(x, fmt))
    df['stay_time'] = (df['stay_time'].dt.total_seconds() / 3600).apply(float)
    return df

fmt = "%Y-%m-%d %H:%M:%S"
df_train = calculate_stay_time(df_train, fmt=fmt)
df_train['stay_time'].head()
```

```python
df_val = calculate_stay_time(df_val, fmt=fmt)
df_val['stay_time'].head()
```

```python
def determine_lengthened_stay(df):
    df['lengthened_stay'] = df['stay_time'].apply(lambda x: x > 24)
    return df

df_train = determine_lengthened_stay(df_train)
df_train['lengthened_stay'].head()
```

```python
df_train['lengthened_stay'].value_counts(normalize=True)
```

```python
df_val = determine_lengthened_stay(df_val)
df_val['lengthened_stay'].value_counts(normalize=True)
```
